```python
import requests
import xml.etree.ElementTree as et
import bs4
from lxml import html
import pprint

url = "https://open.kci.go.kr/po/openapi/openApiSearch.kci"
api_code = "articleSearch"
api_key = "77670960"
search_keyword = "Mechanical properties of recycled fine glass aggregate concrete under uniaxial loading"

# URL 파라미터 설정
params = {
    'apiCode': api_code,
    'key': api_key,
    'title': search_keyword
}

# GET 요청 보내기
response = requests.get(url, params=params)

if response.status_code == 200:
    # 요청이 성공했을 때 처리
    print("성공적으로 요청을 보냈습니다.")
    contents = response.text

    # BeautifulSoup을 사용하여 XML 파싱
    xml_obj = bs4.BeautifulSoup(contents, 'xml')

    # article-id 값을 추출
    article_id_tag = xml_obj.find('articleInfo', {'article-id': True})
    if article_id_tag:
        article_id = article_id_tag['article-id']
        print("article-id 값:", article_id)
    else:
        print("article-id를 찾을 수 없습니다.")

else:
    # 요청이 실패했을 때 처리
    print("요청 실패. 상태 코드:", response.status_code)
```

```python

url = "https://open.kci.go.kr/po/openapi/openApiSearch.kci"
api_code = "articleDetail"
api_key = "77670960"
control_number  = 'ART002018473'

# URL 파라미터 설정
params = {
    'apiCode': api_code,
    'key': api_key,
    'id': control_number
}

# GET 요청 보내기
response = requests.get(url, params=params)

try:
    # 응답 확인
    response.raise_for_status()
    
    # 요청이 성공했을 때 처리
    print("성공적으로 요청을 보냈습니다.")
    contents = response.text
    # 이제 contents에는 요청에 대한 응답 데이터가 들어 있습니다.
    print(contents)

except requests.exceptions.HTTPError as err:
    # 요청이 실패했을 때 처리
    print("요청 실패. 상태 코드:", response.status_code)
    print("상세 메시지:", err)
```
